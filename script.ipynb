# [dbo].[Cliente_Create]

```sql
/****** Object:  StoredProcedure [dbo].[Cliente_Create]    Script Date: 7/8/2022 23:11:07 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[Cliente_Create]
	   @RUC varchar(20),
	   @RAZON_SOCIAL varchar(20),
	   @FECHA_CONSTITUCION	date,
	   @ESTADO bit
	 
AS
BEGIN
INSERT INTO [dbo].[CLIENTES]
           ([RUC]
           ,[RAZON_SOCIAL]
           ,[FECHA_CONSTITUCION]
           ,[ESTADO])
    VALUES (
	   @RUC,
	   @RAZON_SOCIAL,
	   @FECHA_CONSTITUCION,
	   @ESTADO)
 

END
GO
```

# [dbo].[Cliente_Update]

```sql
/****** Object:  StoredProcedure [dbo].[Cliente_Update]    Script Date: 7/8/2022 23:11:07 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[Cliente_Update]
	   @RUC varchar(20),
	   @ESTADO bit
	 
AS
BEGIN
UPDATE [dbo].[CLIENTES] SET ESTADO = @ESTADO WHERE RUC = @RUC

END
GO
```

# [dbo].[Facturas_Insertar]

```sql
/****** Object:  StoredProcedure [dbo].[Facturas_Insertar]    Script Date: 7/8/2022 23:11:07 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[Facturas_Insertar]
	   @RUC varchar(20),
	   @IMPORTE INT
	 
AS
BEGIN
INSERT INTO [dbo].[FACTURAS]
           ([FECHA]
           ,[IMPORTE_TOTAL]
           ,[RUC])
     VALUES
           (GETDATE()
           ,@IMPORTE
           ,@RUC)
END
GO
```

# [dbo].[Facturas_Total]

```sql
/****** Object:  StoredProcedure [dbo].[Facturas_Total]    Script Date: 7/8/2022 23:11:07 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[Facturas_Total]
	   @RUC varchar(20),
	   @Mensaje varchar(50) OUTPUT
	 
AS
BEGIN
SET @Mensaje = CONVERT(VARCHAR,COALESCE((SELECT SUM(IMPORTE_TOTAL) FROM FACTURAS WHERE RUC = @RUC),0))

END
GO
```
